```python
sp["label2"] = sp["label2"].map(lambda x: str(x).replace("\n","").replace("\r","").replace(".", "").replace("no answer", "neutral"))
grouped = sp.groupby(sp.topic)
dfs = []
for _,g in grouped:
    dfs.append(g)
```

```python
answers={}
for topic in root.findall('topic'):
    query = topic.find("description").text
    stance = topic.find("stance").text

    if stance=="unhelpful": ## ADAPTAR ESTO A VARIOS AÑOS
        answers[query.rstrip()] = "no"
    else:
        answers[query.rstrip()] = "yes"
answers
```

```python
### USER'S LAZY BEHAVIOR

hit = 0
fail = 0
n = len(dfs)
ps = []
unanswered = True
for df in dfs:
    p = 0
    for label in df["label2"]:
        if label=="neutral":
            p+=1
            continue
        else:
            unanswered = False
            topic = df.iloc[0,:]["topic"]
            truth = answers[topic]
            p+=1
            if truth==label:
                hit+=1
            else:
                fail+=1
            break
    if unanswered:
        print("Unanswered:", df.iloc[0,:]["topic"])
    unanswered = True
    ps.append(p)
```

```python
statistics.mean(ps)
```

```python
perc_hit = hit/n ## el porcentaje de aciertos parando en el primero
perc_fail = fail/n
perc_neut = 1 - (perc_hit + perc_fail)
esfuerzo_mediano = statistics.median(ps)
```

### User critic (3 no-neutrales -> majority)

```python
#### 2021

import pandas as pd

#### Google
sp = pd.read_csv('stance_prediction_google_2021.csv', names=["topic", "link", "passage", "answer", "label", "label2"])
sp
```

```python
sp["label2"] = sp["label2"].map(lambda x: str(x).replace("\n","").replace("\r","").replace(".", "").replace("no answer", "neutral"))
grouped = sp.groupby(sp.topic)
dfs = []
for _,g in grouped:
    dfs.append(g)
dfs[0]
```

```python
import xml.etree.ElementTree as ET
root = ET.parse("/home/marcos.fernandez.pichel/PhD/TREC/trec-working-lab/evaluation/misinfo-resources-2021/topics/misinfo-2021-topics.xml").getroot() ## open queries file
answers={}
for topic in root.findall('topic'):
    query = topic.find("description").text
    stance = topic.find("stance").text

    if stance=="unhelpful": ## ADAPTAR ESTO A VARIOS AÑOS
        answers[query.rstrip()] = "no"
    else:
        answers[query.rstrip()] = "yes"
answers
```

```python
len(answers)
```

```python
#### USER'S CRITIC BEHAVIOR

hit = 0
fail = 0
n = len(dfs)
ps = []
for df in dfs:
    p = 0
    non_neutral = 0
    labels = []
    for label in df["label2"]:
        if label=="neutral":
            p+=1
            continue
        else:
            non_neutral +=1
            labels.append(label) ## añadimos la respuesta no-neutral para después hacer un voto mayoritario
            p+=1
            if non_neutral==3:
                topic = df.iloc[0,:]["topic"]
                truth = answers[topic]
                if truth==max(labels): ### majority vote 
                    hit+=1
                else:
                    fail+=1
                break
    
    #### Si llega hasta aquí y el len es menor que 3, puede ser que todos hayan sido neutrales, que haya encontrada 1 o 2
    if len(labels) < 3:
        if(len(set(labels))==1): ### si la respuesta o 2 respuestas recolectadas son coincidentes
            topic = df.iloc[0,:]["topic"]
            truth = answers[topic]
            if truth==max(labels): ### majority vote 
                hit+=1
            else:
                fail+=1
            
    ps.append(p)
```

```python
statistics.mean(ps)
```

```python
import statistics

perc_hit = hit/n ## el porcentaje de aciertos parando en el primero
perc_fail = fail/n
perc_neut = 1 - (perc_hit + perc_fail)
esfuerzo_mediano = statistics.median(ps)
```
