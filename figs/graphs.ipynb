```python
import matplotlib.pyplot as plt
import numpy as np

# Datos de la gráfica
models = ['d-002', 'd-003', 'ChatGPT', 'GPT-4', 'Llama 3', 'MedLlama 3', 'Flan T5']
contexts = ['no context', 'non-expert', 'expert']
colors = ['#A8DADC', '#457B9D', '#1D3557']

# Valores para cada contexto y modelo
values = {
    'd-002': [0.72, 0.40, 0.36],
    'd-003': [0.76, 0.62, 0.80],
    'ChatGPT': [0.68, 0.56, 0.70],
    'GPT-4': [0.68, 0.66, 0.68],
    'Llama 3': [0.72, 0.78, 0.8],
    'MedLlama 3': [0.8, 0.82, 0.8],
    'Flan T5': [0.44, 0.54, 0.64]
}

# Crear la gráfica de barras
fig, ax = plt.subplots()

width = 0.2  # Ancho de las barras
x = np.arange(len(models))  # Posición en el eje x para cada modelo

# Añadir las barras para cada contexto
for i, context in enumerate(contexts):
    context_values = [values[model][i] for model in models]
    ax.bar(x + i * width, context_values, width, label=context, color=colors[i])

# Añadir etiquetas y leyenda
ax.set_ylabel('Accuracy', fontsize=16)

# ax.set_title('Model Comparison by Context')
ax.set_xticks(x + width)
ax.set_xticklabels(models, fontsize=16)
ax.legend(loc='lower left', fontsize=16)

# Eliminar el grid
ax.grid(False)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Mostrar la gráfica
plt.tight_layout()
plt.xticks(rotation=35)
plt.yticks(fontsize=16)

plt.savefig('./zeroshot-2021.png', bbox_inches='tight')
plt.show()
```
